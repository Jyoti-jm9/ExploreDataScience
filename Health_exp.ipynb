```python
import pandas as pd
#read the dataset
data = pd.read_csv('/content/insurance.csv')

print(data.head())
print(data.describe())
```

```python
#check for null value
#data.isnull()
print(data.isnull().sum())
```

**No NULL value** is present in the dataset

```python
#encoding
data.replace({'sex':{'male' : 0 , 'female' : 1 , 'other' : 2}} , inplace= True)
data.replace({'smoker' : {'yes' : 1 , 'no' : 0}} , inplace = True)
data.replace({'region' : {'northwest' : 0 , 'northeast' : 1 , 'southwest' : 2 , 'southeast' : 3}} , inplace = True)
data
```

```python
target = data.charges
X = data.drop(columns=['charges'])
X
```

```python
#Spiliting into test and training sets
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X , target , test_size = 0.1 ,   random_state = 9)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression().fit(X_train , Y_train)
prediction = model.predict(X_test)
r2 = r2_score(Y_test , prediction);
mse = mean_squared_error(Y_test , prediction)
print("R-squured : " , r2);
print("mse : "  , mse);
```

```python
#create a dataframe actual vs predicted
results_data = pd.DataFrame({
    'Actual' : Y_test ,
    'Predicted' : prediction
})
results_data
```
